```python
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas

def create_pdf(report_filename):
    c = canvas.Canvas(report_filename, pagesize=letter)
    width, height = letter

    # Title
    c.setFont("Helvetica-Bold", 16)
    c.drawString(100, height - 50, "Weekly Goal Tracking Report")

    # Short Summary
    summary = """
    Key Findings:
    - Goal Success: The overall success rate for the week was 64%, with a notable increase in productivity by 14% compared to the previous week.
    - Daily Performance: The goal completion varied across days, with high success on Wednesday, Thursday, and Monday.
    - Self and Coach Ratings: There was some alignment between the teen’s self-assessment (6/10) and the coach’s rating (5/10).
    """
    c.setFont("Helvetica", 12)
    text_object = c.beginText(100, height - 100)
    text_object.textLines(summary)
    c.drawText(text_object)

    # Add images (closer packing, no overlapping)
    c.drawImage("goals_completion_bar_chart.png", 50, height - 450, width=500, height=300)  # First image
    c.drawImage("goal_progress_line_chart.png", 50, height - 800, width=500, height=300)    # Second image

    # Save PDF
    c.save()

# Create the PDF report
create_pdf("goal_tracking_report.pdf")
```
